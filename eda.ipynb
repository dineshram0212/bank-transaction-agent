```python
"What is the amount I have spent on Uber in the last 5 months?"

# Filter for Uber transactions in the last 5 months
uber_txns = df[
    (df['merchant'].str.contains('UBER', na=False, case=False)) & 
    (pd.to_datetime(df['txn_date']) >= pd.to_datetime('now') - pd.DateOffset(months=5))
]

# Sum up the Uber transactions
uber_total = uber_txns['amt'].sum()

print(f"Total spent on Uber in last 5 months: ${uber_total:.2f}")
```
